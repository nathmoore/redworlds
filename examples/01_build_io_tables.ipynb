# Building the Baseline IO Tables

This notebook shows how to:

1. Download EXIOBASE and load it with pymrio
2. Inspect the IO table structure
3. Calculate baseline emissions — the starting point for every player's world

**Before you start:** follow the instructions in [data/README.md](../data/README.md)
to download EXIOBASE and configure `config/config.toml`.

## 1. Load configuration

```python
# TODO: replace with redworlds.config.load_config() once implemented (GitHub issue #1)
import tomllib
from pathlib import Path

config_path = Path("..") / "config" / "config.toml"
with open(config_path, "rb") as f:
    cfg = tomllib.load(f)

exiobase_path = Path(cfg["data"]["exiobase_path"])
print(f"EXIOBASE path: {exiobase_path}")
print(f"Exists: {exiobase_path.exists()}")
```

## 2. Load EXIOBASE with pymrio

```python
import pymrio

# Load a specific year — adjust as needed
# TODO: parameterise the year (GitHub issue #5)
mrio = pymrio.load_exiobase3(path=exiobase_path)
print(mrio)
```

## 3. Calculate the full IO system

```python
# This calculates Z, L, S, M, D, and all satellite accounts
mrio.calc_all()
print("Calculation complete.")
print(f"Regions: {list(mrio.get_regions())[:5]} ...")
print(f"Sectors: {list(mrio.get_sectors())[:5]} ...")
```

## 4. Inspect baseline emissions

EXIOBASE includes satellite accounts for GHG emissions (and many others).
This is the baseline — what the world emits before any player action.

```python
# TODO: extract total GHG using get_sector_emissions() once implemented (GitHub issue #9)
# For now, inspect the satellite accounts directly
if hasattr(mrio, "emissions"):
    print(mrio.emissions.S.head())
else:
    print("Satellite accounts available:", list(mrio.get_extensions()))
```

## 5. Using the built-in test IO (no EXIOBASE needed)

If you don't have EXIOBASE yet, you can explore the IO table structure using
pymrio's built-in test world. All Red Worlds unit tests use this.

```python
test_mrio = pymrio.load_test()
test_mrio.calc_all()
print("Test IO system:")
print(f"  Regions: {list(test_mrio.get_regions())}")
print(f"  Sectors: {list(test_mrio.get_sectors())}")
print()
print("Z matrix (intermediate demand):")
test_mrio.Z
```
