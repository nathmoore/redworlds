# Sample Red Carbon Payloads

This notebook shows the JSON payloads that the Red Carbon WordPress front end
will send to Red Worlds, and exercises each action type against the pymrio
test IO world.

Use this for manual testing and integration smoke-testing before the full
WordPress → Red Worlds pipeline is connected.

No EXIOBASE files are required — all examples use `pymrio.load_test()`.

```python
import json

import pymrio

# Load and calculate the test IO world
mrio = pymrio.load_test()
mrio.calc_all()

print("Test IO world loaded.")
print(f"  Regions: {list(mrio.get_regions())}")
print(f"  Sectors: {list(mrio.get_sectors())}")
```

## Sample payloads

These are the JSON structures that Red Carbon will insert into the job queue.
Copy-paste these to test the API manually once the server endpoint is up.

```python
# ── BUILD payload ──────────────────────────────────────────────
# Player builds 10 offshore wind farms in Europe over 5 years.
# Budget and build_years come from the Decarbonator Deck output.

build_payload = {
    "action": "build",
    "player_id": "player_001",
    "region": "Europe and Central Asia",
    "technology": "offshore_wind",
    "budget": 450_000,  # currency units in the IO system
    "build_years": 5,
    "current_year": 1,
}

print("BUILD payload:")
print(json.dumps(build_payload, indent=2))
```

```python
# ── SWAP payload ───────────────────────────────────────────────
# Player swaps 32% of replaceable gas heating to heat pumps in Europe.
# pct_rollout is the Deck output as a fraction of the max_replaceable_fraction.

swap_payload = {
    "action": "swap",
    "player_id": "player_001",
    "region": "Europe and Central Asia",
    "from_technology": "gas_heating",
    "to_technology": "heat_pumps",
    "pct_rollout": 0.32,  # fraction in [0.0, 1.0]
}

print("SWAP payload:")
print(json.dumps(swap_payload, indent=2))
```

```python
# ── REDUCE payload ─────────────────────────────────────────────
# Player lowers thermostats — 18% reduction of the reducible share of
# residential heating in Europe.

reduce_payload = {
    "action": "reduce",
    "player_id": "player_001",
    "region": "Europe and Central Asia",
    "sector": "residential_heating",
    "pct_reduction": 0.18,  # fraction in [0.0, 1.0]
}

print("REDUCE payload:")
print(json.dumps(reduce_payload, indent=2))
```

## Sample result (what Red Worlds sends back)

This is the structure written to `result_json` in the job queue record.
Red Carbon reads this to update the player's display and drive the animation.

```python
# ── Example result JSON ────────────────────────────────────────
# This is illustrative — the real structure will evolve as the engine is built.

example_result = {
    "player_id": "player_001",
    "action": "reduce",
    "year": 1,
    "status": "ok",
    "emissions": {
        "before_Mt_CO2eq": 142.3,
        "after_Mt_CO2eq": 134.7,
        "reduction_pct": 5.3,
    },
    "animation_steps": [
        # Intermediate values Red Carbon can display during the 'calculating' sequence
        # e.g. sector-by-sector emissions ticks, rebalancing steps
        # Structure TBC — placeholder for now
        {"step": "load_io_table", "value": None},
        {"step": "apply_reduction", "value": -0.18},
        {"step": "recalculate_emissions", "value": 134.7},
    ],
}

print("Example result JSON:")
print(json.dumps(example_result, indent=2))
```
